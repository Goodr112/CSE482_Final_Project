```python
import pandas as pd
import random
import math
import numpy as num

original_data = pd.read_csv('NYPD_Shooting_Incident_Data__Historic.csv')
original_data.head()
```

```python
acceptable_years = ['2013','2014','2015','2016','2017','2018','2019']
crime_dict = {}
poverty_level = []
row_counter = 0
for row in range(len(original_data.index)):
    current_row = original_data.iloc[row]
    if current_row['OCCUR_DATE'][-4:] in acceptable_years:
        crime_dict[row_counter] = current_row
        if current_row['BORO'] == 'BRONX':
            poverty_level.append(27)
        elif current_row['BORO'] == 'BROOKLYN':
            poverty_level.append(19.6)
        elif current_row['BORO'] == 'MANHATTAN':
            poverty_level.append(13.2)
        elif current_row['BORO'] == 'QUEENS':
            poverty_level.append(17.8)
        else:
            poverty_level.append(16.8)
        row_counter += 1
crime_data = pd.DataFrame(crime_dict)
crime_data = crime_data.transpose()
crime_data['POVERTY_LEVEL'] = poverty_level
print(crime_data.shape)
crime_data.head()
```

```python
sample_size = 6000
data_size = len(crime_data.index)
crime_sample = {}
for row in range(data_size):
    current_row = crime_data.iloc[row]
    if len(crime_sample) < sample_size:
        crime_sample[row] = current_row
    else:
        rand_val = random.random()
        if rand_val <= data_size/row:
            rand_num = random.randint(0,5999)
            crime_sample[rand_num] = current_row
sample = pd.DataFrame(crime_sample)
sample = sample.transpose()
print(sample.shape)
```

```python
for row in range(len(sample.index)):
    current_row = sample.iloc[row]
    if current_row['DAY_OF_THE_WEEK'] == 'Sunday':
        current_row['DAY_OF_THE_WEEK'] = 1
    elif current_row['DAY_OF_THE_WEEK'] == 'Monday':
        current_row['DAY_OF_THE_WEEK'] = 2
    elif current_row['DAY_OF_THE_WEEK'] == 'Tuesday':
        current_row['DAY_OF_THE_WEEK'] = 3
    elif current_row['DAY_OF_THE_WEEK'] == 'Wednesday':
        current_row['DAY_OF_THE_WEEK'] = 4
    elif current_row['DAY_OF_THE_WEEK'] == 'Thursday':
        current_row['DAY_OF_THE_WEEK'] = 5
    elif current_row['DAY_OF_THE_WEEK'] == 'Friday':
        current_row['DAY_OF_THE_WEEK'] = 6
    elif current_row['DAY_OF_THE_WEEK'] == 'Saturday':
        current_row['DAY_OF_THE_WEEK'] = 7
    if current_row['BORO'] == 'BROOKLYN':
        current_row['BORO'] = 1
    elif current_row['BORO'] == 'BRONX':
        current_row['BORO'] = 2
    elif current_row['BORO'] == 'QUEENS':
        current_row['BORO'] = 3
    elif current_row['BORO'] == 'MANHATTAN':
        current_row['BORO'] = 4
    elif current_row['BORO'] == 'STATEN ISLAND':
        current_row['BORO'] = 5
sample.head()
```
